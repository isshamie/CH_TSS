## First merge with tissues, then separate by unique and overlapping sets

```python
from tss.utils.Homer import merge_peaks, annotate_peaks
```

```python
import pandas as pd
import os
from os.path import join, basename
import glob
import shutil
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
samples = pd.read_csv("../parameters/samples.tsv", sep="\t", index_col=0)
samples
```

## Directory names

```python
import yaml
parameter_file = "../parameters/params.yaml"
with open(parameter_file,'r') as f:
    doc = yaml.load(f)

print(doc)
gtf = doc["annotation_gtf"]
ref_fa = doc["ref_fa"]

results = doc["results"]
os.chdir(results)
```

```python
out_dir = "csRNA_pipeline/"
csRNA_dir = join(out_dir, 'csRNA_peaks')


homer_path =  "/data2/resources/software/homer/v4.11_10242019/bin/"
os.environ['PATH'] = f"{homer_path}:{os.environ['PATH']}"
!which homer
```

## Outdirs

```python
tissues_csRNA_dir = join(out_dir,"csRNA_peaks_tissues")
unique_csRNA_dir =  join(out_dir,"csRNA_peaks_tissues.merged")

if not os.path.exists(tissues_csRNA_dir):
    os.mkdir(tissues_csRNA_dir)
    
if not os.path.exists(unique_csRNA_dir):
    os.mkdir(unique_csRNA_dir)
    
merged_f = join(unique_csRNA_dir, 'csRNA_peaks.tissues.merge')
anno_f = merged_f + '.anno'
seq_f = anno_f + '.fa'
```

## Merge tissues

```python
for tis, df in samples.groupby("Tissue"):#samples.loc[samples["Tissue"].isin(["FemaleReproductive", "MiscOrgans"])].groupby("Tissue"):
    print(tis)
    files = glob.glob(join(csRNA_dir,"*"+tis+"*.tss.txt" ))
    if len(files) != 0:
        out_f = join(tissues_csRNA_dir, tis+'.tss.txt.merge')
#         mat_f = out_f + ".matrix"
#         venn_f = out_f + ".venn"
#         cmd = f"mergePeaks -d given {' '.join(files)} -matrix {mat_f} -venn {venn_f} > {out_f}"
#         #print(files)
#         print(cmd)
#         os.system(cmd)
        
        merge_peaks(input_files=files, output_file=out_f, dist='given', type_merge='')
```

## Merge over all tissues 

```python
# for tis, df in samples.groupby("Tissue"):#samples.loc[samples["Tissue"].isin(["FemaleReproductive", "MiscOrgans"])].groupby("Tissue"):
#     print(tis)
files = glob.glob(join(tissues_csRNA_dir,"*.tss.txt.merge" ))
if len(files) != 0:   
    #out_f = join(unique_csRNA_dir, 'csRNA_peaks.tissues.merge')
#     mat_f = out_f + ".matrix"
#     venn_f = out_f + ".venn"
#     cmd = f"mergePeaks -d given {' '.join(files)} -matrix {mat_f} -venn {venn_f} > {out_f}"
#     #print(files)
#     print(cmd)
#     os.system(cmd)

    merge_peaks(input_files=files, output_file=merged_f, dist='given', type_merge='')
```

### Annotate

```python
annotate_peaks(peak_file=merged_f,output_file=anno_f,ref_fa=ref_fa,annotation=gtf )
```

### Extract sequences

```python
cmd = 'homerTools extract {f_in} {ref_fa} -fa > {seq_f}'.format(f_in=anno_f,ref_fa=ref_fa,seq_f=seq_f)
!{cmd}
```
