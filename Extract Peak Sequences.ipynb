```python
import pickle
import numpy as np
import pandas as pd
```

```python
gene_df_1kb_1kb = pickle.load(open("Results/gene_df_1kb_1kb.p",'rb'))
```

```python
gene_df_1kb_1kb.head()
```

```python

#Remove empty values to None
#gene_df['prom_1kb_1kb'] = gene_df['prom_1kb_1kb'].apply(lambda x: return None if x == '' else return ) == None
gene_df_1kb_1kb = gene_df_1kb_1kb[(~gene_df_1kb_1kb['tss_bin_-1000_1000'].isnull())]

### Save promoter peak set
inds = [item for sublist in list(gene_df_1kb_1kb['tss_bin_-1000_1000']) for item in sublist]
```

```python
anno_peaks = pd.read_csv('Results/peaks_with_tss_and_gene.tsv',sep='\t',index_col=0)
```

```python
anno_peaks_short = anno_peaks[anno_peaks.index.isin(inds)]
```

```python
anno_peaks_short.to_csv('Results/promoter_1kb_1kb.tsv',sep='\t')
```
