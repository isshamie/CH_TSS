## Load tss meta

```python
eTSS_meta = pd.read_csv(eTSS_meta_f,sep='\t', index_col=0)
stable_eTSS = pd.read_csv(join(save_dir,"eTSS_merged_stable_peaks.tsv.merge"),sep='\t', index_col=0)
stable_eTSS

stable_eTSS_inds = list()
for i in stable_eTSS[~(stable_eTSS.isnull().any(axis=1))][eTSS_bed_f].apply(lambda x:np.array(x.split(','))):
    for j in i:
        stable_eTSS_inds.append(j)

eTSS_meta['cs'] = eTSS_meta['cs'].fillna('')
for ind, val in tqdm(eTSS_meta.iterrows()):
    if val['cs']=='':
        continue
    elif ind in stable_eTSS_inds:
        eTSS_meta.at[ind,'cs'] = val['cs'] + '1'
    else:
        eTSS_meta.at[ind,'cs'] = val['cs'] + '0'
eTSS_meta
```

```python
eTSS_meta = eTSS_meta.drop("Gene ID",axis=1) 
```

```python
eTSS_meta.to_csv(join(save_dir, os.path.basename(eTSS_meta_f)), sep='\t')
```
