## Run truncatedSVD 

```python
from sklearn.decomposition import TruncatedSVD

from scipy import stats
```

### First normalize the data

```python
peak_by_motif_norm = stats.zscore(peak_by_motif)
```

## Run with n=4 components

```python
%%time

svd = TruncatedSVD(n_components=4)
svd.fit(peak_by_motif_norm)

print(svd.explained_variance_ratio_)  
print(len(svd.explained_variance_ratio_))

print(svd.explained_variance_ratio_.sum())  

print(svd.singular_values_) 

print(svd.components_.shape)
```

### Plot first 2 axes

```python
pc_axes_peak_by_motif = svd.transform(peak_by_motif)
```

```python
pd.DataFrame(pc_axes_peak_by_motif[:,:2]).plot.hexbin(x=0,y=1,gridsize=30)
```

```python
pd.DataFrame(pc_axes_peak_by_motif[:,:2]).plot.hexbin(x=0,y=1,gridsize=10)
```

```python
### Look at tata box weightings
svd.components_[:,np.flatnonzero(peak_by_motif.columns.str.contains("tata"))[0]]
```

### Use 10 dimensions and cluster

```python
%%time

svd_10 = TruncatedSVD(n_components=10)
svd_10.fit(peak_by_motif_norm)

print(svd_10.explained_variance_ratio_)  
print(len(svd_10.explained_variance_ratio_))

print(svd_10.explained_variance_ratio_.sum())  

print(svd_10.singular_values_) 

print(svd_10.components_.shape)

### Plot first 2 axes

pc_axes_peak_by_motif_10 = svd_10.transform(peak_by_motif)
```

### Cluster with new axes

```python
%%time
clusterer = hdbscan.HDBSCAN()
clusterer.fit(pc_axes_peak_by_motif_10)
```
