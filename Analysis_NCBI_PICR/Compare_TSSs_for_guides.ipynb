## For the p0 peaks move, the positions 1 bp for the bed file
1. If the strand is +, move end - 1 
2. If the strand is -, move start + 1

## Generate flanking +/- 1kb sequences too


```python
#extract_peak_sequences()
bed_f = "Results/target_genes/ncbi_103_allGenes/ncbi_103_allGenes.tsv"
out = "Results/target_genes/ncbi_103_allGenes/ncbi_103_allGenes.fa"
cmd = "homerTools extract {bed_f} {ref_fa} -fa > {out}".format(bed_f=bed_f,ref_fa=ref_fa,out=out)
print(cmd)
!{cmd}
```

```python
peaks = pd.read_csv(bed_f,sep="\t")
for ind,val in peaks.iterrows():
    if "p0@" in val["ID"]:
        if val["Strand"] is "+":
            peaks.at[ind,"End"] = val["End"] - 1
        else:
            peaks.at[ind,"Start"] = val["Start"] + 1
    else:
        center = int((val["End"]+val["Start"])/2)
        peaks.at[ind,"Start"] = center
        peaks.at[ind,"End"] = center + 1
        
peaks["Start"] -= 1000
peaks["End"] += 1000
peaks.to_csv("Results/target_genes/ncbi_103_allGenes/flank_ncbi_103_allGenes.tsv",sep="\t",index=False)


#extract_peak_sequences()
flank_f = "Results/target_genes/ncbi_103_allGenes/flank_ncbi_103_allGenes.tsv"
out = "Results/target_genes/ncbi_103_allGenes/ncbi_103_allGenes.fa"
cmd = "homerTools extract {bed_f} {ref_fa} -fa > {out}".format(bed_f=flank_f,ref_fa=ref_fa,out=out)
print(cmd)
!{cmd}
peaks
```

## Determine which genes are missing to blast against genome

```python
genes_not_in_new_anno_f = "Results/target_genes/genes_not_in_new_anno"
if not os.path.exists(genes_not_in_new_anno_f):
    os.mkdir(genes_not_in_new_anno_f)
genes_not_in_new_anno_f = genes_not_in_new_anno_f + "genes_not_in_new_anno.txt"
```

```python
not_in_new_annotation = set()
for i in secmExpr_genes:
    for j in secmExpr_genes[i]:
        if j not in tss_meta_genes:
            not_in_new_annotation.add(j)

with open(genes_not_in_new_anno_f, 'w') as f:
    for item in not_in_new_annotation:
        f.write("%s\n" % item)

not_in_new_annotation
```
