```python
parameter_file = 'params/params.yaml'
import yaml
import sys
import glob
import os
import pandas as pd
with open(parameter_file,'r') as f:
    doc = yaml.load(f)

#p = dic2obj(**doc)

data_folder = doc['data_folder']
tissues = doc['tissues'].split(',')
sys.path.append(doc['pipeline_path'])
ref_fa = doc['ref_fa']
annotation=doc['annotation_gtf']
```

```python
meta_files = pd.read_csv("../raw_metaSamples.tsv",sep='\t',index_col=0)
meta_files
```

```python
### Merge based on any sample file (so replicates are treated separately)
#### And annotate
all_peak_merge_files = []
for t in tissues:
    for s in ["GROCap","START"]:
        curr_folder = os.path.join(data_folder,t,s,"f03_tags/trim*")
        all_peak_merge_files = all_peak_merge_files + glob.glob(curr_folder)
```

```python
all_peak_merge_files
```

```python
copy_files = map(lambda x: os.path.basename(x).replace("trim_","").replace(".fastq",""),all_peak_merge_files)

copy_files = map(lambda x: os.path.basename(x).replace("STARTinput","startinput"),copy_files)
copy_files = map(lambda x: os.path.join('f03_tags/',x), copy_files)
if not os.path.exists("f03_tags"):
    os.mkdir("f03_tags")

for i in zip(all_peak_merge_files,copy_files):
    cmd = "cp -r %s %s" % (i[0] ,i[1])
    print(cmd)
    !{cmd}
```
