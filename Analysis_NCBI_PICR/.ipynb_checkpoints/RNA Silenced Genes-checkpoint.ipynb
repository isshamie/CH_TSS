```python
cho_rna_file = doc["cho_rna_file"]
cho_rna = pd.read_csv(cho_rna_file,sep="\t",index_col=0)
cho_rna = cho_rna.iloc[:,1:]
#cho_rna = np.log10(cho_rna+1)
cho_rna
```

### Load picr_final start site file, since the RNA-seq was mapped to that genome

```python
old_mrna = pd.read_csv("/data/isshamie/genome/picr_final/mRNA_final.peak",sep="\t")
old_mrna
```

### Combine gene_ids with similar genes by taking the median expression

```python
cho_rna_gnames = cho_rna.groupby(old_mrna.set_index(["gene_id"],)["gene"].to_dict()).median()
```

```python
cho_rna_gnames
```

## Analysis

```python
(((cho_rna_gnames < 1).sum(axis=1) / cho_rna_gnames.shape[1]) > 0.9).sum()
```

```python
1.0*(((cho_rna_gnames < 1).sum(axis=1) / cho_rna_gnames.shape[1]) > 0.9).sum()/cho_rna_gnames.shape[0]
```

```python
plt.hist((cho_rna_gnames < 1).sum(axis=1) / cho_rna_gnames.shape[1])
```

```python
gene_df = pickle.load(open("Results/tss_annotation/gene_df_02.p","rb"))#, encoding='latin')
gene_df
```

```python
cho_rna_gnames
```

```python
len(set(gene_df.index.str.upper()).intersection(set(cho_rna_gnames.index)))
```

```python
cho_rna_gene_overlap = (cho_rna_gnames.loc[set(gene_df.index.str.upper()).intersection(set(cho_rna_gnames.index))])
plt.hist((cho_rna_gene_overlap < 1).sum(axis=1) / cho_rna_gene_overlap.shape[1])
print("Number of genes with less than 1 TPM in greater than 90% of samples:", 
     (((cho_rna_gene_overlap < 1).sum(axis=1) / cho_rna_gene_overlap.shape[1]) > 0.9).sum())

print("Fraction of genes with less than 1 TPM in greater than 90% of samples:", 
     1.0*(((cho_rna_gene_overlap < 1).sum(axis=1) / cho_rna_gene_overlap.shape[1]) > 0.9).sum()/cho_rna_gene_overlap.shape[0])
```
