```python
tissue_df = gene_tissue.loc[:,~(gene_tissue.columns.isin(['CHO']))]
cho_df = gene_tissue.loc[:,(gene_tissue.columns.isin(['CHO']))]
```

```python
#######################################
def cho_cpmThresh_nTis(tissue_df,cho_df,cpm=1,n_tis=1):
    '''Parameters:
        tissue_df: The genexsample expression matrix. 
        A tissue could have more than one sample, and if at least one is present then it is ok.
    '''
    
    in_cho = cho_df[((cho_df>cpm).any(axis=1))].index.values
    in_cho = list(map(lambda x: x.upper(), in_cho))
    not_in_cho = tissue_df[(np.sum(tissue_df>cpm,axis=1)>=n_tis) & ((cho_df<=cpm).all(axis=1))].index.values
    not_in_cho = list(map(lambda x: x.upper(), not_in_cho))
    return in_cho, not_in_cho
```

```python
in_cho, not_in_cho = cho_cpmThresh_nTis(tissue_df,cho_df,cpm=1,n_tis=1)
print(len(in_cho))
print(len(not_in_cho))
print(in_cho)
```

```python
in_cho_c0_t1, not_in_cho_c0_t1 = cho_cpmThresh_nTis(tissue_df,cho_df,cpm=0,n_tis=1)
print(len(in_cho_c0_t1))
print(len(not_in_cho_c0_t1))
```

```python
in_cho_c1_t3, not_in_cho_c1_t3 = cho_cpmThresh_nTis(tissue_df,cho_df,cpm=1,n_tis=3)
print(len(in_cho_c1_t3))
print(len(not_in_cho_c1_t3))
```

```python
import xml.etree.ElementTree as et
```

#### Convert from id to symbol

```python
gene2refseq = pd.read_csv(join(supplemental,'gene2refseq'),sep='\t',header=0,
                         usecols=['GeneID','Symbol'])
```

```python
unique_genes = gene2refseq["Symbol"].unique()
print(len(unique_genes))
```

```python
len(gene2refseq["Symbol"].str.upper().unique())
```

```python
er_f = join(supplemental,'KEGG_Pathways','cge04141.xml')

er = set()
tree = et.parse(er_f )
root = tree.getroot()
for child in root:
    #print child.tag, child.attrib
    if 'name' in child.attrib and 'cge:' in child.attrib['name'] : 
        er.add(child.attrib['name'].split(':')[1].split()[0]) #Take the first cge:number 
        

er_symbol = set()
er_symbol_dict = dict()
for i in er:
    for j in gene2refseq[gene2refseq['GeneID'] == int(i)]['Symbol']:
        er_symbol_dict[j.upper()] = i
        er_symbol.add(j)
        
er_symbol = set(map(lambda x: x.upper(),er_symbol))


print(er_symbol.intersection(set(in_cho)))

print(er_symbol.intersection(set(not_in_cho)))
```

### Prepare for KEGG Map

```python
for i in (er_symbol.intersection(set(not_in_cho))):
    print(er_symbol_dict[i] + ' orange,black')

for i in (er_symbol.intersection(set(in_cho))):
    print(er_symbol_dict[i] + ' yellow,black')
```

### Apoptotic genes

```python
## Load in apoptotic genes
apop1_f  = join(supplemental,'KEGG_Pathways','cge04210.xml')
apop2_f = join(supplemental,'KEGG_Pathways','cge04215.xml')

apop1 = set()
tree = et.parse(apop1_f )
root = tree.getroot()
for child in root:
    #print child.tag, child.attrib
    if 'name' in child.attrib and 'cge:' in child.attrib['name'] : 
        apop1.add(child.attrib['name'].split(':')[1].split()[0]) #Take the first cge:number 
        
apop2 = set()
tree = et.parse(apop2_f)
root = tree.getroot()
for child in root:
    #print child.tag, child.attrib
    if 'name' in child.attrib and 'cge:' in child.attrib['name']: 
        apop2.add(child.attrib['name'].split(':')[1].split()[0])

apop1_symbol = set()
for i in apop1:
    for j in gene2refseq[gene2refseq['GeneID'] == int(i)]['Symbol']:
        apop1_symbol.add(j)

apop2_symbol = set()
for i in apop2:
    for j in gene2refseq[gene2refseq['GeneID'] == int(i)]['Symbol']:
        apop2_symbol.add(j)

print(len(apop2))
print(len(apop1))

apop1_symbol = set(map(lambda x: x.upper(),apop1_symbol))
apop2_symbol = set(map(lambda x: x.upper(),apop2_symbol))

apop = apop1_symbol.union(apop2_symbol)

print(apop.intersection(set(not_in_cho)))

print(apop.intersection(set(in_cho)))
```

## Create table of pathways and in/not in CHO

```python
kegg_df = pd.DataFrame(columns=["Pathway", "KEGG ID", "Number of Genes in Pathway", 
                      "Number of Genes in CHO", "Number of Genes Rescued in at least 1 BioSample"])
#                       "Number of Genes Rescued in at least 1 BioSample","Number of Genes Rescued in at least 3 BioSample",
#                      "Number of Genes Rescued in at least 1 BioSample","Number of Genes Rescued in all BioSamples"],
#             index={"ER", "Secretory Machinery", "DNA Repair Genes", "Apoptotic", "N-Glyosylation",})

kegg_df

kegg_ids = {'ER':'cge04141',"N-Glycosylation":"cge00510", 
           'Apoptosis 1': 'cge04210','Apoptosis 2':'cge04215',
           'Base Excision Repair': 'cge03410', 'Nuclear Repair':'cge03420'}

for k in kegg_ids:
    print(k)
    curr_kegg_f = join(supplemental, "KEGG_Pathways", kegg_ids[k]+".xml")
    #nglyco_f = join(supplemental,'KEGG_Pathways/cge00510_nglycan_biosyn.xml')

    curr_pathway = set()
    tree = et.parse(curr_kegg_f)
    root = tree.getroot()
    for child in root:
        #print child.tag, child.attrib
        if 'name' in child.attrib and 'cge:' in child.attrib['name'] : 
            curr_pathway.add(child.attrib['name'].split(':')[1].split()[0]) #Take the first cge:numbnglyco 


    curr_symbol = set()
    curr_symbol_dict = dict()
    for i in curr_pathway:
        for j in gene2refseq[gene2refseq['GeneID'] == int(i)]['Symbol']:
            curr_symbol_dict[j.upper()] = i
            curr_symbol.add(j)

    curr_symbol = set(map(lambda x: x.upper(),curr_symbol))

    
    curr_in_cho = len(curr_symbol.intersection(set(in_cho)))
    curr_not_in_cho = len(curr_symbol.intersection(set(not_in_cho)))
    #print(curr_symbol.intersection(set(in_cho)))
    #print(curr_symbol.intersection(set(not_in_cho)))
    
    kegg_df = pd.concat((kegg_df, pd.DataFrame({"Pathway": k,
                                                "KEGG ID": kegg_ids[k], 
                                                "Number of Genes in Pathway": len(curr_symbol),
                                                "Number of Genes in CHO": curr_in_cho,
                                                "Number of Genes Rescued in at least 1 BioSample": curr_not_in_cho,
                                                }, 
                                               index=[kegg_ids[k]])),sort=False)
```
