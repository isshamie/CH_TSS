## Annotate based on totalRNA

```python
peak_anno = pd.read_csv('Results/peaks_with_tss_and_gene.tsv',index_col=0,sep='\t')
peak_anno.dropna(axis=1,how='all',inplace=True)
peak_anno.head()
```

```python
tss_totalRNA = pd.read_csv('/data/isshamie/genome/totalRNA_start.peak',sep='\t',index_col=0)

### Need to create an extra column for tss called 'actual_start'. 
### This is to take care of issue of a gene starting at the End when it's on the minus strand
tss_totalRNA['actual_start'] = tss_totalRNA.apply( lambda x: x['Start'] if x['Strand'] == '+' else x['End'],axis=1)

tss_totalRNA.head()

```

#### Calculate distance of each peak to nearest RNA peak

```python
peak_anno['Nearest tss_totalRNA'] = ''
peak_anno['Distance to tss_totalRNA'] = np.infty
peak_anno['isSameStrand RNA'] = False

for ind,val in tqdm.tqdm_notebook(peak_anno.iterrows()):
    filt = tss_totalRNA[val['Chr'] == tss_totalRNA['Chr']]    
    peak_start = (val['Start']+val['End'])/2
    filt = filt[val['Strand'] == tss_totalRNA['Strand']]
    #print(filt)
    if len(filt) != 0:
        #print(filt2)
        filt2 = np.abs(filt['actual_start'] - peak_start).argmin()
        peak_anno.set_value(ind,'Nearest tss_totalRNA',filt2)
        
        ## Get distance to nearest gene. If on - strand, tss_totalRNA 'End' is the beginning
        if tss_totalRNA.loc[filt2,'Strand'] == '+':
            peak_anno.set_value(ind,'Distance to tss_totalRNA',peak_start - tss_totalRNA.loc[filt2,'actual_start'])
        elif tss_totalRNA.loc[filt2,'Strand'] == '-':
            peak_anno.set_value(ind,'Distance to tss_totalRNA', tss_totalRNA.loc[filt2,'actual_start']-peak_start)
        
        #peak_anno.set_value(ind,'isSameStrand',val['Strand'] == tss_totalRNA.loc[filt2,'Strand'])
        
        
        
        
        
        
```
